## Solution

We have an LCG imlementation with random addtitive terms. Namely:

$s_n = a * s_{n-1} + r_{n} \pmod{p}$, $p =$ next_prime $(2^{1024})$

````python
p = next_prime(2^1024)
p
````

In this implementation $r_n \in [-2^{768}, 2^{768}]$ which is much less than 1024.

Once again we are having lattice problem and a recurring [paper](https://eprint.iacr.org/2020/1506.pdf) page 30.

We can transform our system of equations, to have only $r_n$ entries:

\begin{aligned}
& s_2 = a * s_{1} + r_{2} \pmod{p} \\
& s_3 = a * s_{2} + r_{3} \pmod{p} \\
& ... \\
& s_{20} = a * s_{19} + r_{20} \pmod{p}
\end{aligned}

Using first two equations we can obtain: $s_3 * s_1 - s_2 * s_2 = a * s_2 * s_1 - a * s_1 * s_2 + r_3 * s_1 - r_2 * s_2 = r_3 * s_1 - r_2 * s_2$

$r_3 - r_2 * s_2 * s_1^{-1} - (s_3 - s_2 * s_2 * s_1^{-1}) = 0 \pmod{p}$

Similarly:

$r_4 - r_2 * s_3 * s_1^{-1} - (s_4 - s_2 * s_1^{-1} * s_3) = 0 \pmod{p}$

etc

We end up with 18 equations of the form

$r_n + u * r_2 + v = 0 \pmod{p}$, $r_i \in [-2^{768}, 2^{768}]$

Then we construct a Lattice:

\begin{align*}
L &= \begin{pmatrix}
p &   & &  \\
 &  p  & &  \\
 & & \ddots &  \\
 & & & p & \\
u_1 & u_2 & \cdots & u_{18} & 1 & \\
v_1 & v_2 & \cdots & v_{18} & 0 & 2^{768}\\
\end{pmatrix}, \\
\end{align*}

And find an element in LLL'd matrix such that all the entries are in $[-2^{768}, 2^{768}]$

````python
s = [0] + samples
us = [(-s[i] * pow(s[1], -1, p)) % p for i in range(2, len(s) - 1)]
vs = [(-(s[i + 1] - s[2] * pow(s[1], -1, p) * s[i])) % p for i in range(2, len(s) - 1)]

maxx = 2^768
MMM = len(us)
M = Matrix(MMM + 2)
M.set_block(0, 0, identity_matrix(MMM) * p)
M.set_block(MMM, 0, Matrix(us))
M.set_block(MMM + 1, 0, Matrix(vs))
M[MMM, MMM] = 1
M[MMM + 1, MMM + 1] = maxx
````

````python
for l in M.LLL():
    if all(abs(x) <= maxx for x in l):
        rds1 = l[:-1]
        break

a1 = (s[2] - rds1[-1]) * pow(s[1], -1, p) % p # recovering multiplier "a" from first equation
print(a1.to_bytes(40, 'big'))
````
